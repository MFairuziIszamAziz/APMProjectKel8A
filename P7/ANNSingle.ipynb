```python
import tensorflow as tf
import pandas as pd
import numpy as np 
```

```python
dataset = pd.read_csv('data.csv')
dataset
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
y =y.reshape(len(y), 1)
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan,strategy='mean')
imputer.fit(X[:, 5:6])
imputer.fit(y)
X[:, 5:6] = imputer.transform(X[:, 5:6])
y=imputer.transform(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc_x_1 = StandardScaler()
sc_y = StandardScaler()
sc_x_2 = StandardScaler()
X_train[:, 0:3]= sc_x_1.fit_transform(X_train[:,0:3])
X_train[:, 4:13]= sc_x_2.fit_transform(X_train[:,4:13])
y_train = sc_y.fit_transform(y_train)
```

```python
ann = tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(500, input_shape=(13,), activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(500, activation='relu'))
ann.add(tf.keras.layers.Dense(500, activation='relu'))
ann.add(tf.keras.layers.Dense(500, activation='relu'))
ann.add(tf.keras.layers.Dense(500, activation='relu'))
ann.add(tf.keras.layers.Dense(500, activation='relu'))
ann.add(tf.keras.layers.Dense(500, activation='relu'))
ann.add(tf.keras.layers.Dense(500, activation='relu'))
ann.add(tf.keras.layers.Dense(500, activation='relu'))
ann.add(tf.keras.layers.Dense(500, activation='relu'))
ann.add(tf.keras.layers.Dense(500, activation='relu'))
ann.add(tf.keras.layers.Dense(500, activation='relu'))
ann.add(tf.keras.layers.Dense(500, activation='relu'))
ann.add(tf.keras.layers.Dense(500, activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(1, activation='linear'))
```

```python
ann.compile(optimizer='Adam', loss='mean_squared_error', metrics=['accuracy'])
```

```python
ann.fit(X_train, y_train, epochs=1000, verbose=1)
```

```python
X_test[:, 0:3]= sc_x_1.transform(X_test[:,0:3])
X_test[:, 4:13]= sc_x_2.transform(X_test[:,4:13])
```

```python
y_pred = sc_y.inverse_transform(ann.predict(X_test))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
